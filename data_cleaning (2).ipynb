```python
df_eda = df_eda.rename(columns = {'Price':'Price_in_million_euro'})
```

```python
def remove_million(df, column_name):

    df[column_name] = df[column_name].str.replace('milliomn', '')
    return df

df_eda = remove_million(df_eda, 'Price_in_million_euro')
```

```python
df_eda_str = df_eda[
    (df_eda['Price_in_million_euro'] == 'Loan') |
    (df_eda['Price_in_million_euro'] == 'undisclosed') |
    (df_eda['Price_in_million_euro'] == 'Free') |
    (df_eda['Price_in_million_euro'] == 'Swap') |
    (df_eda['Price_in_million_euro'] == 'co-ownership') |
    (df_eda['Price_in_million_euro'] == 'Undisclosed') |
    (df_eda['Price_in_million_euro'] == 'P/Exchange') |
    (df_eda['Price_in_million_euro'] == 'loan')
]
```

```python
df_eda_num = df_eda[
    (df_eda['Price_in_million_euro'] != 'Loan') &
    (df_eda['Price_in_million_euro'] != 'undisclosed') &
    (df_eda['Price_in_million_euro'] != 'Free') &
    (df_eda['Price_in_million_euro'] != 'Swap') &
    (df_eda['Price_in_million_euro'] != 'co-ownership') &
    (df_eda['Price_in_million_euro'] != 'Undisclosed') &
    (df_eda['Price_in_million_euro'] != 'P/Exchange') &
    (df_eda['Price_in_million_euro'] != 'loan')
]
```
