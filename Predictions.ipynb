```python
position_mapping = {
      'ST': 'ATT',
      'CDM': 'MID',
      'CB': 'DEF',
      'LM': 'MID',
      'LW': 'ATT',
      'RW':'ATT',
      'RB':'DEF',
      'CM':'MID',
      'GK':'DEF',
      'LWB':'DEF',
      'RM':'MID',
      'CF':'ATT',
      'CAM':'MID',
      'RWB':'DEF',
      'LB':'DEF'
  }

df_all['Position'] = df_all['Position'].replace(position_mapping)
```

```python
latin_american_countries = [
    "Uruguay",
    "Costa Rica",
    "Chile",
    "Argentina",
    "Colombia",
    "Ecuador",
    "Brazil",
    "Mexico",
    "Paraguay",
    "Honduras",
    "Peru",
    "Venezuela",
    "Panama",
    "Mexico"
]

balkan_countries = [
    "Bulgaria",
    "FYR Macedonia",
    "Croatia",
    "Serbia",
    "Slovenia",
    "Albania",
    "Kosovo",
    "Romania",
    "Montenegro",
    "Bosnia and Herzegovina"
]

african_countries = [
    "Congo",
    "Senegal",
    "Zambia",
    "Nigeria",
    "Togo",
    "Morocco",
    "South Africa",
    "Ghana",
    "Cameroon",
    "Ivory Coast",
    "Kenya",
    "Algeria",
    "Gabon",
    "DR Congo",
    "Tunisia",
    "Egypt",
    "Benin",
    "Gambia",
    "Republic of Tanzania",
    "Zimbabwe",
    "Guinea",
    "Sierra Leone",
    "Mali",
    "Libya",
    "Burkina Faso",
    "Angola"
]

scandinavian_countries = [
    "Norway",
    "Sweden",
    "Denmark",
    "Iceland",
    "Finland"
]

britannia_countries = [
    'Wales',
    'Scotland'
]

def map_to_latin_america(country):
  if country in latin_american_countries:
    return "Latin America"
  else:
    return country

def map_to_balkan(country):
  if country in balkan_countries:
    return "Balkan"
  else:
    return country

def map_to_african(country):
  if country in african_countries:
    return "African"
  else:
    return country

def map_to_scandinavian(country):
  if country in scandinavian_countries:
    return "Scandinavian"
  else:
    return country

def map_to_britania(country):
  if country in britannia_countries:
    return "Britania"
  else:
    return country
```

```python
df_all['Country'] = df_all['Country'].apply(map_to_latin_america)
df_all['Country'] = df_all['Country'].apply(map_to_african)
df_all['Country'] = df_all['Country'].apply(map_to_balkan)
df_all['Country'] = df_all['Country'].apply(map_to_scandinavian)
df_all['Country'] = df_all['Country'].apply(map_to_britania)

country_counts = df_all['Country'].value_counts()
values_to_replace = country_counts[country_counts < 50].index
df_all.loc[df_all['Country'].isin(values_to_replace), 'Country'] = 'Others'
```

<h1>Modelling

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from catboost import CatBoostClassifier

from sklearn.metrics import accuracy_score, classification_report
```

```python
class_names = ['Premier League', 'Seria A', 'Bundesliga', 'La Liga']
models = []
model_names = []
```

```python
model_svm = SVC()
model_svm.fit(X_train, y_train)

models.append(model_svm)
model_names.append('SVM')

model_random_frst = RandomForestClassifier()
model_random_frst.fit(X_train, y_train)
models.append(model_random_frst)
model_names.append('Random Forest')


model_gr_boost = GradientBoostingClassifier()
model_gr_boost.fit(X_train, y_train)
models.append(model_gr_boost)
model_names.append('Gradient Boosting')

model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(X_train, y_train)
models.append(model_knn)
model_names.append('kNN')

model_catboost = CatBoostClassifier(
    iterations=1000,
    learning_rate=0.1,
    depth=6,
    loss_function='MultiClass',
    verbose=0,  # To avoid printing the training process
    random_state=42
)

model_catboost.fit(X_train, y_train)

models.append(model_catboost)
model_names.append('catboost')
```

```python
predictions_train = []; predictions_test = []
for i in range(len(models)):
  model = models[i]
  y_pred_train = model.predict(X_train)
  predictions_train.append(y_pred_train)

  y_pred = model.predict(X_test)
  predictions_test.append(y_pred)
```

```python
from collections import Counter

train_final_answer = []
for i in range(X_train.shape[0]):
    ans = []
    for pred in predictions_train:
        ans.append(pred[i].item() if np.isscalar(pred[i]) else tuple(pred[i]))

    counter = Counter(ans)
    most_common_element, frequency = counter.most_common(1)[0]
    train_final_answer.append(most_common_element)
```

```python
test_final_answer = []
for i in range(X_test.shape[0]):
    ans = []
    for pred in predictions_test:
        ans.append(pred[i].item() if np.isscalar(pred[i]) else tuple(pred[i]))

    counter = Counter(ans)
    most_common_element, frequency = counter.most_common(1)[0]
    test_final_answer.append(most_common_element)
```

```python
accuracy = accuracy_score(y_train, train_final_answer)
print(f"Train Accuracy: {accuracy:.4f}")
```
