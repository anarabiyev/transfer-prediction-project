```python
def shorten_name(name):
    """
    Shortens a name by extracting the first letter of the first name
    and appending it to the last name.

    Example:
    Leo Messi -> L. Messi
    """

    parts = name.split()
    first_initial = parts[0][0] + "."
    shortened_name = " " + first_initial + " " + parts[-1]
    return shortened_name

df_transfers['Name'] = df_transfers['Name'].apply(shorten_name)
```
